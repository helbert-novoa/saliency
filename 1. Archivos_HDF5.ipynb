### Librerias

```python
import h5py
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Lista de carpetas 

```python
#Se carga el archivo con extensión HDF, alojado en la carpeta específica, y se procede a listar las tablas correspondientes a cada uno de los estudios contenidos en él.

hdf = h5py.File("D:/Documentos/etdb_v1.0.hdf5", mode = "r")
list(hdf.keys())
```

### Variables de una carpeta especifica

```python
#De la lista de estudios se selecciona uno en concreto y se determina las variables que conforman la tabla. Este proceso se debe efetuar para cada estudio.

estudio = hdf.get("Webtask")
variables = list(estudio.keys())
variables
```

### Dataframe de la carpeta especifica

```python
# Del estudio seleccionado se construye un dataframe

df = pd.DataFrame(estudio.values()).T
df.columns = variables
df
```

### Dataframe Age Study

```python
#Para el estudio Age Study se consolida el datafreme final

df_age = df[df["valid"]==1]
df_age["study"]="Age study"
df_age= df_age[["study", "category", "filenumber", "x", "y"]]
df_age
```

### Dataframe Baseline

```python
#Para el estudio Baseline se consolida el datafreme final

df_base = df[df["eye"]==1]
df_base["study"]="Baseline"
df_base= df_base[["study", "category", "filenumber", "x", "y"]]
df_base
```

### Dataframe Bias

```python
#Para el estudio Bias se consolida el datafreme final

df_bias = df[df["mirror"]==0]
df_bias = df_bias[df_bias["on_image"]==1]
df_bias["study"]="Bias"
df_bias= df_bias[["study", "category", "filenumber", "x", "y"]]
df_bias
```

### Dataframe Memory I

```python
#Para el estudio Memory I se consolida el datafreme final

df_memory = df[df["eye"]==1]
df_memory = df_memory[df_memory["iteration"]==1]
df_memory = df_memory[df_memory["on_image"]==1]
df_memory["study"]="Memory I"
df_memory= df_memory[["study", "category", "filenumber", "x", "y"]]
df_memory
```

### Dataframe Webtask

```python
#Para el estudio Webtask se consolida el datafreme final

df_web = df[df["condition"]==1]
df_web = df_web[df_web["eye"]==1]
df_web = df_web[df_web["on_image"]==1]
df_web["x"] =df_web["x"]-4
df_web["y"] =df_web["y"]-86
df_web["study"]="Webtask"
df_web= df_web[["study", "category", "filenumber", "x", "y"]]
df_web
```

```python
### Dataframe Final
```

```python
#Se procede a concatenar los dataframes de los estudios seleccionados en un único dataframe

df_final = pd.concat([df_age, df_base, df_bias, df_memory, df_web], axis=0).reset_index()
df_final
```

```python
#fijaciones por estudio

freq = df_final["category"].value_counts().reset_index()
freq["categoria"] = ["Urbano", "Fractal", "Natural", "Ruido", "Web"]
freq_final = pd.DataFrame({"categoria": freq["categoria"], "ejemplos": freq["category"]}, columns= ["categoria", "ejemplos"])
plt.style.use('ggplot')
plt.figure(figsize=(10,8))
g =sns.barplot(freq_final["categoria"], freq_final["ejemplos"])
for index, row in freq_final.iterrows():
    g.text(row.name,row.ejemplos, round(row.ejemplos,2), color='black', ha="center")
plt.title("Distribución del número de fijaciones por categoría")
plt.xlabel("Categoría de análisis")
plt.ylabel("Número de fijaciones")
plt.xticks(rotation=30)
plt.show()
```

```python
#Tabla de frecuencias. Número de fijaciones por estudio

freq = df_final["category"].value_counts().reset_index()
freq["categoria"] = ["Urbano", "Fractal", "Natural", "Ruido", "Web"]
freq_final = pd.DataFrame({"categoria": freq["categoria"], "ejemplos": freq["category"]}, columns= ["categoria", "ejemplos"])
freq_final
```

### Exportación del dataframe a CSV

```python
#Exportacion del dataframe a un archivo separado por comas

df_final.to_csv("D:/Documentos/dataset.csv")
print("Archivo generado con exito!")
df_final.shape
```
